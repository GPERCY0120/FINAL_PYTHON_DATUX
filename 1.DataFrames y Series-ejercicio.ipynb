# Editar el indice (index)

```python
# Cambia el indice a la columna "fecha_inicio".
df2_Avenger=df_Avenger.set_index("fecha_inicio")
df2_Avenger
```

# Ordenar el indice

```python
# Ordena el √≠ndice de forma descendiente
df2_Avenger.sort_values(by=["fecha_inicio"],ascending=[False])
```

# Resetear el indice
Este metodo es un proceso, No retorna nada

```python
# Resetea el √≠ndice
df2_Avenger.reset_index(drop=True, inplace=True)
df2_Avenger
```
